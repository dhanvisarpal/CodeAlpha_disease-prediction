```python
import numpy as np
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', None)
```

```python
# Load CSV file
df = pd.read_csv("breast cancer dataset.csv")

#print first 5 rows 
print(df.head())

#print last 5 rows 
print("\n", df.tail())
```

```python
#dataset information 
df.info()
```

```python
#statistical info
df.describe()
```

```python
#missing values
print("missing value if any :",df.isnull().sum())

#duplicate if any 
print("\nduplicate value: ",df.duplicated().sum())

```

```python
df.drop(columns=["id", "Unnamed: 32"], errors="ignore", inplace=True)
```

```python
#counting diagnosis column 
df["diagnosis"].replace({
    "M": "Malignant",
    "B": "Benign"
}).value_counts()
```

```python
df["diagnosis"] = df["diagnosis"].map({
    "M": 1,
    "B": 0
})
```

```python
df["diagnosis"].value_counts()
```

```python
#occurence of m and b in diagnosis column
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="diagnosis", data=df)
plt.title("Diagnosis Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
features = [
    "radius_mean",
    "texture_mean",
    "perimeter_mean",
    "area_mean",
    "concavity_mean",
    "concave points_mean",
    "radius_worst",
    "area_worst"
]
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
for ax, feature in zip(axes.flatten(), features):
    sns.histplot(df[feature], kde=True, ax=ax)
    ax.set_title(feature)
plt.tight_layout()
plt.show()
```

```python
features = [
    "radius_mean",
    "texture_mean",
    "perimeter_mean",
    "area_mean",
    "concavity_mean",
    "concave points_mean",
    "radius_worst",
    "area_worst"
]
fig, axes = plt.subplots(2, 4, figsize=(18, 8))
for ax, feature in zip(axes.flatten(), features):
    sns.boxplot(x=df[feature], ax=ax)
    ax.set_title(feature)
plt.tight_layout()
plt.show()
for col in features:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = ((df[col] < lower) | (df[col] > upper)).sum()

    print(f"{col}: {outliers} outliers")
```

```python
df.to_csv("processed_breast cancer.csv", index=False)
print("saved successfully")
```
